## Eliminar datos faltantes y nulos:

En la documentación para el fix se menciona que se hace uso de **-1** para los valores null o fuera del rango, por lo que nuestra tarea es limpiar estos datos

puesto que se está usando el valor -1 como indicador para los valores nulos o faltantes debemos buscar todas la filas
que contienen dicho valor:

```python
missing_values = accidents[accidents.eq(-1).any(1)]
missing_values
# el método .eq(n) busca el número n, mientras que con .any(axis) le indicamos dónde buscar
# en Pandas axis=0 es usado para los renglones y axis=1 para las columnas, por lo que la línea
# de arriba busca el valor -1 en barriendo todas las columnas, es decir, si encuentra un -1 devuelve
# esa fila
```

podemos observar que tenemos un total 97500 filas que contienen al menos un valor -1, con esto ya encontramos los valores faltantes, en este caso elegiremos remover los datos faltantes puesto que la cantidad de datos es alta,
con esto recordemos que el data frame original cuenta con 207410 filas de las cuales 97500 cuentan con datos faltantes, por lo que nuestro data freme sin datos faltantes debe contar con 207410 - 97500 = **109910**

```python
accidents = accidents[accidents.ne(-1).all(1)]
# en el ejemplo anterior buscamos cualquier fila que contuviera el valor -1, ahora buscamos todas las
# filas que no contienen el valor -1, usamos ne: not equal en lugar de eq y all en lugar de any
```

```python
len(accidents) # podemos comprobar que el número de filas que obtenemos corresponde a la resta que calculamos
```

Hasta este punto tenemos nuestro data frame libre de datos faltantes

```python
accidents.head()
```
